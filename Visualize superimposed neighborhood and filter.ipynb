```python
import os
from visualizations import show_3d_filters,show_3d_neighborhoods,weight_logo_3d
from utilities import dataset_utils
import predict_features
import numpy as np
model_name = 'ScanNet_PAI_noMSA_0'
dataset_name = 'BCE_fold1'
list_origins = dataset_utils.read_labels('datasets/BCE/labels_fold1.txt')[0]

filter_specificities = show_3d_filters.calculate_filter_specificities(
    model_name,
    dataset_name = dataset_name,
    dataset_origins = list_origins,
    biounit=False,
    ncores=4,
    only_atom=False

)
```

```python
pdb = '7jvb'
model = 0
chain = 'A'


dictionary_features = predict_features.predict_features('%s_%s'%(pdb,chain),model=model_name)

dictionary_epitopes = predict_features.predict_features('%s_%s'%(pdb,chain),model=model_name,
                                                       layer=None )

dictionary_attention = predict_features.predict_features('%s_%s'%(pdb,chain),model=model_name,
                                                       layer='attention_layer' )

dictionary_epitopes_preaverage = predict_features.predict_features('%s_%s'%(pdb,chain),model=model_name,
                                                       layer='node_output' )



```

```python
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
matrix_features = np.stack(list(dictionary_features.values()),axis=0)
matrix_percentile_features = show_3d_filters.activity2percentile(matrix_features, filter_specificities['filter_activity']['percentiles'] )


vector_epitopes = np.array(list(dictionary_epitopes.values()))
vector_attention = np.array(list(dictionary_attention.values()))
vector_preaverage_epitopes = np.array(list(dictionary_epitopes_preaverage.values()))
vector_preaverage_epitopes = -(vector_preaverage_epitopes[:,1] - vector_preaverage_epitopes[:,0] )

resids = [key[2] for key in dictionary_features.keys() ]
L = len(resids)

fig, ax = plt.subplots(figsize=(16,8))
sns.heatmap( matrix_features.T ,ax = ax); 
plt.show()


fig, ax = plt.subplots(figsize=(16,8))
sns.heatmap( matrix_percentile_features.T ,ax = ax,vmin=95,vmax=100); 
plt.show()



fig, ax = plt.subplots(figsize=(16,8))
ax.plot( vector_epitopes )
ax2 = ax.twinx()
ax2.plot(vector_attention,c='gray')
ax.set_xticks( np.arange(L)[::4] )
ax.set_xticklabels( resids[::4],rotation=90 )
ax.grid()
plt.show()





a = 1.0
b = -0.5
vector_preaverage_epitopes2 = 1/(1+np.exp(- a*vector_preaverage_epitopes - b  ) )
    


fig, ax = plt.subplots(figsize=(16,8))
ax.plot( vector_epitopes )
ax2 = ax.twinx()
ax2.plot(vector_preaverage_epitopes2,c='gray')
ax.set_xticks( np.arange(L)[::4] )
ax.set_xticklabels( resids[::4],rotation=90 )
ax.grid()
plt.show()



```

```python
salient_residues, salient_features = np.nonzero(matrix_percentile_features>99.4)
salient_residues = np.array(resids)[salient_residues]

for sr,sf in zip(salient_residues,salient_features):
    print(sr,sf)
```

```python
resnumber = 455
normalized = True


filter_activity = dictionary_features[(model,chain,resnumber)]
filter_activity_percentile = show_3d_filters.activity2percentile(filter_activity, filter_specificities['filter_activity']['percentiles'] )

if normalized:
    top_filters = np.argsort(filter_activity_percentile )[::-1][:5]    
else:
    top_filters = np.argsort(filter_activity )[::-1][:5]
    
for top_filter in top_filters:
    print(top_filter,filter_activity[top_filter],filter_activity_percentile[top_filter])

    
plt.plot(filter_activity); plt.show()
plt.plot(filter_activity_percentile); plt.show()
```
